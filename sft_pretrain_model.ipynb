# **Training a Conversational Model using SFTTrainer**

The SFTTrainer from the TRL (Transformers Reinforcement Learning) library is a powerful tool used for supervised fine-tuning of language models. It enables developers and researchers to refine pre-trained Large Language Models (LLMs) using task-specific datasets in order to improve their performance for targeted applications such as conversational AI, question answering, summarization, and instruction-following systems.

In this notebook, we will explore how to train a conversational model using the **SFTTrainer** framework together with the Hugging Face Transformers ecosystem. Instead of building a language model from scratch, supervised fine-tuning allows us to adapt an already pre-trained model to better understand and generate human-like conversations.

The notebook will guide us through the complete fine-tuning pipeline, including:
- Loading a pre-trained conversational model
- Preparing and tokenizing a dialogue dataset
- Configuring training parameters
- Training the model using SFTTrainer
- Evaluating the model’s conversational responses

By completing this exercise, we will gain hands-on experience with modern transformer-based conversational AI training workflows and understand how supervised fine-tuning is applied in real-world natural language processing applications.

```python
import os
os.environ["PYTHONUTF8"] = "1"
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, DataCollatorForLanguageModeling
from trl import SFTTrainer
from trl.trainer import SFTConfig
from transformers import pipeline
```

```python
import torch
print(torch.__version__)
print(torch.cuda.is_available())
```

```python
dataset = load_dataset("timdettmers/openassistant-guanaco")
```

```python
dataset_train = dataset['train']
dataset_train[0]
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device.type
```

```python
model = AutoModelForCausalLM.from_pretrained("facebook/opt-350m")
tokenizer = AutoTokenizer.from_pretrained("facebook/opt-350m")

model.to(device)
```

### **Data Collator**

```python
### in modern LLM fine tuning we dont use collator function anymore
```

#### **SFT Config, SFT Trainer**

```python
training_args = SFTConfig(
    output_dir="/tmp",
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=10,
    save_strategy='epoch',
    max_length=1024,
    do_eval=True
)
```

```python
trainer = SFTTrainer(
    model= model,
    args=training_args,
    train_dataset=dataset_train,
    processing_class=tokenizer
)
```

### **Model Pipeline Design**

```python
pipe = pipeline(task='text-generation', tokenizer=tokenizer, model=model, device=device)
```

```python
result = pipe('''Can you write a short introduction about the relevance of the term "monopsony" in economics? Please use examples related to potential monopsonies in the labour market and cite relevant research.''')
```
